## Setup

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
import numpy as np
```

## Load data

```python
data = load_digits()
```

## build TSNE model

```python
n_components = 2
model_pca = PCA(n_components=n_components)
model_tsne = TSNE(n_components=n_components)
model_lda = LDA(n_components=n_components)
```

## Implement TSNE model

```python
points_pca = model_pca.fit_transform(data.data)
points_tsne = model_tsne.fit_transform(data.data)
points_lda = model_lda.fit_transform(data.data, data.target)
```
